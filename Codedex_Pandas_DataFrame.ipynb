```python
import pandas as pd

data = {
    "Title": [
        "Spirited Away", "Your Name", "Interstellar", "Inception", "The Matrix",
        "Notting Hill", "Crazy Rich Asians", "Her", "Weathering With You", "About Time"
    ],
    "Genre": [
        "Anime", "Anime", "Sci-Fi", "Sci-Fi", "Sci-Fi",
        "Rom-Com", "Rom-Com", "Rom-Com", "Anime", "Rom-Com"
    ],
    "Year": [
        2001, 2016, 2014, 2010, 1999,
        1999, 2018, 2013, 2019, 2013
    ],
    "Director": [
        "Hayao Miyazaki", "Makoto Shinkai", "Christopher Nolan", "Christopher Nolan", "The Wachowskis",
        "Roger Michell", "Jon M. Chu", "Spike Jonze", "Makoto Shinkai", "Richard Curtis"
    ],
    "Rating": [
        8.6, 8.4, 8.7, 8.8, 8.7,
        7.2, 6.9, 8.0, 7.5, 7.8
    ]
}

movies = pd.DataFrame(data)
movies
```

```python
#add new column "watched"
movies['watched'] = ['True', 'True', 'True', 'False', 'False', 'False', 'False', 'False', 'True', 'False']
movies
```

```python
#sort by rating
rating = movies.sort_values(by='Rating', ascending=False)
rating
```

```python
#filter for watching-list
watching_list = movies[movies['watched'] == 'False']
watching_list
```
